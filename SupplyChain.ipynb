# **Exploratory Data Analysis (EDA)**
Exploratory Data Analysis (EDA) merupakan proses awal untuk memahami karakteristik, pola, dan hubungan antar variabel sebelum melakukan analisis lebih lanjut atau pemodelan.

Exploratory Data Analysis meliputi:
1. Pengecekan informasi data.
2. Pengecekan jumlah baris dan kolom.
3. Deteksi duplikasi data.
4. Menghapus variabel yang tidak relevan.
5. Pengecekan missing value dan memutuskan untuk menghapus atau mengganti value yang hilang.
6. Mendeteksi Outlier.
